```python
import json, re, pathlib, pandas as pd
from io import StringIO
from pathlib import Path

# ── 0) helper: smart filename parser ────────────────────────────────────────
def parse_fname(stem: str):
    """
    Accepts p1_scene1_coded   or  p1_scene1_seed
            run01             etc.
    Returns dict(pid=..., scene=...)
    """
    tokens = stem.split("_")
    pid   = next((t for t in tokens if t.startswith("p")),  None)
    scene = next((t for t in tokens if t.startswith("scene")), None)
    task  = next((t for t in tokens if t.startswith("task")),  None)
    return pid, scene or task or "unknown"

# ── 1) universal loader for *any* JSON(L) flavour ───────────────────────────
_BLANK_RE = re.compile(r"\n\s*\n+", re.MULTILINE)

def load_jsonl(folder: Path) -> pd.DataFrame:
    frames = []
    for fn in sorted(folder.glob("*.jsonl")):
        text = fn.read_text("utf-8").strip()
        if not text:
            continue

        # A) try strict JSONL
        try:
            df = pd.read_json(StringIO(text), lines=True)
        except ValueError:
            # B) json array? (starts with “[” and ends with “]”)
            try:
                js = json.loads(text)
                if isinstance(js, list):
                    df = pd.DataFrame(js)
                else:
                    raise ValueError
            except (json.JSONDecodeError, ValueError):
                # C) pretty-printed objects separated by blank lines
                records = []
                for blk in _BLANK_RE.split(text):
                    blk = blk.strip(" ,\n")
                    if not blk:
                        continue
                    try:
                        records.append(json.loads(blk))
                    except json.JSONDecodeError:
                        continue
                df = pd.DataFrame.from_records(records)

        # keep only rows that *have* id & open_code (filters out the header row)
        if {"id", "open_code"}.issubset(df.columns):
            df = df[df["id"].notna() & df["open_code"].notna()]

        pid, scene = parse_fname(fn.stem)
        df["pid"], df["scene"] = pid, scene
        frames.append(df)

    return pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()

# ── 2) paths (assumes variables from your first cell exist) ────────────────
df_human_manual = load_jsonl(HUMAN_DIR / "open-codes" / "IRR" / "manual-coded")
df_human_llm    = load_jsonl(HUMAN_DIR / "open-codes" / "IRR" / "llm-coded")

print("manual rows:", len(df_human_manual))
print("llm-coded rows:", len(df_human_llm))
common = set(df_human_manual.id) & set(df_human_llm.id)
print("overlap on id:", len(common))

# ── 3) human DAT (single json file) ─────────────────────────────────────────
def load_human_dat(path: Path) -> pd.DataFrame:
    """
    Accepts:
      1) a proper JSON array  [ {...}, {...} ]
      2) JSONL  {...}\n{...}
      3) newline-delimited with trailing semicolon  {...};\n{...};
    Returns one tidy DataFrame.
    """
    text = path.read_text("utf-8").strip()
    records = []

    # ------------- 1) try full JSON load ----------------
    try:
        obj = json.loads(text)
        if isinstance(obj, list):
            records = obj
        elif isinstance(obj, dict):           # single participant
            records = [obj]
    except json.JSONDecodeError:
        pass

    # ------------- 2) fallback: split into objects ------ 
    if not records:
        # remove wrapping [ ] if user left them in but the JSON inside is broken
        if text.startswith("[") and text.endswith("]"):
            text = text[1:-1]

        # split on either  "} ;"  OR a blank line
        parts = re.split(r"}\s*;|\n\s*\n+", text)
        for p in parts:
            p = p.strip(" ,;\n\t")
            if not p:
                continue
            # ensure each chunk ends with } before parsing
            if not p.endswith("}"):
                p += "}"
            try:
                records.append(json.loads(p))
            except json.JSONDecodeError:
                continue    # skip bad chunks

    # ----------- convert to long format -----------------
    rows, step_re = [], re.compile(r"^\s*(\d+)[\.:]\s*(\w+)")
    for rec in records:
        pid = rec.get("participant_id", "UNKNOWN")
        for sc in rec.get("scenes", []):
            scene = str(sc.get("scene", "")).replace("scene", "")
            for step in sc.get("sequence", []):
                m = step_re.match(step)
                if m:
                    rows.append(
                        dict(pid=pid,
                             scene=scene,
                             position=int(m.group(1)),
                             code=m.group(2))
                    )

    return pd.DataFrame(rows)


df_human_dat = load_human_dat(HUMAN_DIR / "DAT" / "participant_DAT.json")

# ── 4) LLM DAT (combine five tasks per buffer) ─────────────────────────────
def load_one_buffer(buf_dir: Path) -> pd.DataFrame:
    recs, step_re = [], re.compile(r"^\s*(\d+)[\.:]\s*(\w+)")
    for task_dir in sorted((buf_dir / "DAT").iterdir()):
        f = task_dir / "participant_DAT.json"
        if not f.exists():
            continue
        data = json.load(open(f, "r", encoding="utf-8"))
        for obj in data:
            run = obj["run_id"]
            for sc in obj["scenes"]:
                scene = sc["scene"].replace("task", "")
                for step in sc["sequence"]:
                    m = step_re.match(step)
                    if m:
                        recs.append({
                            "buffer": buf_dir.name,
                            "run_id": run,
                            "scene": scene,
                            "position": int(m.group(1)),
                            "code": m.group(2)
                        })
    return pd.DataFrame(recs)

agent_dat = {buf.name: load_one_buffer(buf)
             for buf in sorted(LLM_DIR.iterdir()) if (buf / "DAT").exists()}

# ── 5) sanity prints ───────────────────────────────────────────────────────
print("manual rows:", len(df_human_manual))
print("llm-coded rows:", len(df_human_llm))
print("human-DAT rows:", len(df_human_dat))
for k, v in agent_dat.items():
    print(f"{k}: {len(v)} rows")
import json, re, pathlib, pandas as pd
from io import StringIO
from pathlib import Path

# ── 0) helper: smart filename parser ────────────────────────────────────────
def parse_fname(stem: str):
    """
    Accepts p1_scene1_coded   or  p1_scene1_seed
            run01             etc.
    Returns dict(pid=..., scene=...)
    """
    tokens = stem.split("_")
    pid   = next((t for t in tokens if t.startswith("p")),  None)
    scene = next((t for t in tokens if t.startswith("scene")), None)
    task  = next((t for t in tokens if t.startswith("task")),  None)
    return pid, scene or task or "unknown"

# ── 1) universal loader for *any* JSON(L) flavour ───────────────────────────
_BLANK_RE = re.compile(r"\n\s*\n+", re.MULTILINE)

def load_jsonl(folder: Path) -> pd.DataFrame:
    frames = []
    for fn in sorted(folder.glob("*.jsonl")):
        text = fn.read_text("utf-8").strip()
        if not text:
            continue

        # A) try strict JSONL
        try:
            df = pd.read_json(StringIO(text), lines=True)
        except ValueError:
            # B) json array? (starts with “[” and ends with “]”)
            try:
                js = json.loads(text)
                if isinstance(js, list):
                    df = pd.DataFrame(js)
                else:
                    raise ValueError
            except (json.JSONDecodeError, ValueError):
                # C) pretty-printed objects separated by blank lines
                records = []
                for blk in _BLANK_RE.split(text):
                    blk = blk.strip(" ,\n")
                    if not blk:
                        continue
                    try:
                        records.append(json.loads(blk))
                    except json.JSONDecodeError:
                        continue
                df = pd.DataFrame.from_records(records)

        # keep only rows that *have* id & open_code (filters out the header row)
        if {"id", "open_code"}.issubset(df.columns):
            df = df[df["id"].notna() & df["open_code"].notna()]

        pid, scene = parse_fname(fn.stem)
        df["pid"], df["scene"] = pid, scene
        frames.append(df)

    return pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()

# ── 2) paths (assumes variables from your first cell exist) ────────────────
df_human_manual = load_jsonl(HUMAN_DIR / "open-codes" / "IRR" / "manual-coded")
df_human_llm    = load_jsonl(HUMAN_DIR / "open-codes" / "IRR" / "llm-coded")
assert len(df_human_manual) == len(df_human_llm), "Mismatched manual vs LLM rows"

# ── 3) human DAT (single json file) ─────────────────────────────────────────
def load_human_dat(path: Path) -> pd.DataFrame:
    raw  = json.load(open(path, "r", encoding="utf-8"))
    rows = []
    step_re = re.compile(r"^\s*(\d+)[\.:]\s*(\w+)")
    for rec in raw:
        pid = rec["participant_id"]
        for sc in rec["scenes"]:
            scene = sc["scene"].replace("scene", "")
            for s in sc["sequence"]:
                m = step_re.match(s)
                if m:
                    rows.append({
                        "pid": pid,
                        "scene": scene,
                        "position": int(m.group(1)),
                        "code": m.group(2)
                    })
    return pd.DataFrame(rows)

df_human_dat = load_human_dat(HUMAN_DIR / "DAT" / "participant_DAT.json")

# ── 4) LLM DAT (combine five tasks per buffer) ─────────────────────────────
def load_one_buffer(buf_dir: Path) -> pd.DataFrame:
    recs, step_re = [], re.compile(r"^\s*(\d+)[\.:]\s*(\w+)")
    for task_dir in sorted((buf_dir / "DAT").iterdir()):
        f = task_dir / "participant_DAT.json"
        if not f.exists():
            continue
        data = json.load(open(f, "r", encoding="utf-8"))
        for obj in data:
            run = obj["run_id"]
            for sc in obj["scenes"]:
                scene = sc["scene"].replace("task", "")
                for step in sc["sequence"]:
                    m = step_re.match(step)
                    if m:
                        recs.append({
                            "buffer": buf_dir.name,
                            "run_id": run,
                            "scene": scene,
                            "position": int(m.group(1)),
                            "code": m.group(2)
                        })
    return pd.DataFrame(recs)

agent_dat = {buf.name: load_one_buffer(buf)
             for buf in sorted(LLM_DIR.iterdir()) if (buf / "DAT").exists()}

# ── 5) sanity prints ───────────────────────────────────────────────────────
print("manual rows:", len(df_human_manual))
print("llm-coded rows:", len(df_human_llm))
print("human-DAT rows:", len(df_human_dat))
for k, v in agent_dat.items():
    print(f"{k}: {len(v)} rows")
```

## Process data


Next we process exclusions. That is, we remove practice trials, trials where the starting numbers were the same as the practice trial, and trials where the participant said nothing relevant to the task.


```python
from sklearn.metrics import cohen_kappa_score
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

# 1) COHEN’S κ on open‐codes
# ensure the columns exist after the new load
need = {"id","open_code"}
assert need.issubset(df_human_manual.columns), f"manual lacks {need-df_human_manual.columns}"
assert need.issubset(df_human_llm.columns),    f"llm-coded lacks {need-df_human_llm.columns}"

merged = (
    df_human_manual[["id","open_code"]]
)
kappa = cohen_kappa_score(merged["open_code_man"],
                          merged["open_code_llm"])
print(f"Cohen’s κ (open‐code IRR): {kappa:.3f}")

# 2) GRAPH-EDIT-DISTANCE IRR
def normalized_ged(g1, g2):
    raw = next(run_code(g1, g2), None)  if hasattr(run_code, "__call__") \
          else nx.graph_edit_distance(g1, g2)
    denom = max(g1.number_of_nodes()+g1.number_of_edges(),
                g2.number_of_nodes()+g2.number_of_edges(),
                1)
    return raw/denom if raw is not None else 1.0

# attach graphs to the two human DFs
for df in (df_human_manual, df_human_llm):
    df["graph"] = df["lm_code_translation"].apply(
        lambda code: run_code(code) if isinstance(code,str) else None
    )

geds = []
for pid, g_man in zip(df_human_manual["id"], df_human_manual["graph"]):
    g_llm = df_human_llm.loc[df_human_llm["id"]==pid, "graph"].iloc[0]
    if g_man and g_llm:
        geds.append(normalized_ged(g_man, g_llm))
print("Mean normalized GED (IRR):", np.mean(geds))

# 3) ACROSS-BUFFER GED COMPARISON
all_buf_rows = []
for buf, df_buf in agent_dat.items():
    # merge LLM sequences→graphs
    df_buf = df_buf.copy()
    df_buf["graph"] = df_buf["code"].apply(lambda oc: run_code(oc))
    # for each run/scene match human
    for _, row in df_buf.iterrows():
        hmatch = df_human_dat.query(
            "participant_id=='p' + @row.run_id[-2:] and scene==@row.scene"
        )
        if not hmatch.empty:
            all_buf_rows.append({
                "buffer": row.buffer,
                "scene": row.scene,
                "run_id": row.run_id,
                "norm_ged": normalized_ged(
                    hmatch["graph"].iloc[0], row.graph
                )
            })
df_buf_ged = pd.DataFrame(all_buf_rows)

plt.figure(figsize=(6,4))
sns.boxplot(x="buffer", y="norm_ged", data=df_buf_ged,
            order=sorted(df_buf_ged["buffer"].unique()))
plt.xticks(rotation=45)
plt.title("Norm. GED: human vs LLM across buffers")
plt.tight_layout()
plt.show()
```

### Split-half correlation analysis

We compute the trial-level correlations between participants in the think-aloud condition and the no think-aloud condition for three problem groups, and compare these correlations to the split-half correlation for the same problem groups in the no think-aloud condition.

We make this comparison using bootstrapping. Since there were more participants in the think-aloud condition than the no think-aloud condition, it is possible that the split half correlations are higher simply by law of large numbers. To account for this, each iteration of bootstrapping uses only 30 participants' data, to match the amount of data we have in the no think-aloud condition.


```python
num_samples = 10_000
condition_dfs_vp = {}
condition_dfs_novp = {}
comparison_conditions = df_trials_3conditions_noExclusions_proc["CONDITION"].unique()
for condition in comparison_conditions:
    df_cond_vp = df_trials_3conditions_noExclusions_proc.query(
        "CONDITION == @condition"
    )
    condition_dfs_vp[condition] = df_cond_vp
    df_cond_novp = df_trials_control_proc.query("CONDITION == @condition")
    condition_dfs_novp[condition] = df_cond_novp

n_pps_per_group = 10

# compute within correlations
correlations_within_vp = []
for _ in tqdm(range(num_samples)):
    pids1, pids2 = [], []
    for condition in comparison_conditions:
        df_cond = condition_dfs_vp[condition]
        pids = df_cond["pid"].unique()
        chosen_pids = rng.choice(pids, size=2 * n_pps_per_group, replace=False)
        pids1.extend(chosen_pids[:n_pps_per_group])
        pids2.extend(chosen_pids[n_pps_per_group:])

    # sample num_participants_per_sample from each array
    df_vp_sample1 = df_trials_3conditions_noExclusions_proc.query("pid in @pids1")
    df_vp_sample2 = df_trials_3conditions_noExclusions_proc.query("pid in @pids2")
    # get trials df for each participant group
    correlation = compute_item_correlation(df_vp_sample1, df_vp_sample2)
    correlations_within_vp.append(correlation)

correlations_within_vp = np.array(correlations_within_vp)

# compute between correlations
correlations_between_groups = []
for _ in tqdm(range(num_samples)):
    # sample num_participants_per_sample from each array
    pids_vp, pids_novp = [], []
    for condition in comparison_conditions:
        df_cond_vp = condition_dfs_vp[condition]
        df_cond_novp = condition_dfs_novp[condition]
        pids_vp.extend(df_cond_vp["pid"].sample(n_pps_per_group, replace=False))
        pids_novp.extend(df_cond_novp["pid"].sample(n_pps_per_group, replace=False))

    # get trials df for each participant group
    df_vp_sample = df_trials_3conditions_noExclusions_proc.copy().query(
        "pid in @pids_vp"
    )
    df_novp_sample = df_trials_control_proc.copy().query("pid in @pids_novp")
    correlation = compute_item_correlation(df_vp_sample, df_novp_sample)
    correlations_between_groups.append(correlation)

correlations_between_groups = np.array(correlations_between_groups)
```

```python
# Compute the proportion of times within-group correlation is greater than between-group
proportion_greater = np.mean(correlations_within_vp > correlations_between_groups)

# Calculate means
mean_within = np.mean(correlations_within_vp)
mean_between = np.mean(correlations_between_groups)

# get a null distribution
mean_diff = mean_between - mean_within
diffs = correlations_within_vp - correlations_between_groups
null_distribution = diffs - mean_diff

# null_distribution = correlations_within_vp - mean_within
p_value = (np.abs(mean_diff) <= np.abs(null_distribution)).mean()

# Print results
print(f"Mean within-VP correlation: {mean_within:.2f}")
print(f"Mean between-groups correlation: {mean_between:.2f}")
print(f"Proportion of times within > between: {proportion_greater:.3f}")
print(f"mean_diff: {mean_diff:.3f}")
print(f"Permutation test p-value: {p_value:.3f}")
```
